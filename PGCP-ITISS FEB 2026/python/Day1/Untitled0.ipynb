```python
#convert marks to grades
# >90% --> A+
# 80% < A <= 90%
# 70% < B <= 80%
# 60% < C <= 70%
# 50% < D <= 60%
# 40% <= P <= 50%
#<40%  --> F
marks = int(input("Enter your end module marks (out of 40)-->"))
per = marks/40*100
if per > 90:
  print("A+")
elif 80<per <=90:
  print("A")
elif 70<per <=80:
  print("B")
elif 60<per <=70:
  print("C")
elif 50<per <=60:
  print("D")
elif 40<= per <=50:
  print("P")
elif per <40:
  print("F")
```

```python
#take marks of 7 students if all of them have more than 16 then print all pass
#method 1:: C approach
i=0
count=0
while i<7:
  marks = int(input("Enter your marks (out of 40)"))
  if marks < 16:
    print("fail")
    count +=1
  # else:
    # print("all pass")
  i=i+1

if count ==0:
  print("all pass")
```

```python
#method 2:: C approach
i=0
flag=0
while i<7:
  marks = int(input("Enter your marks (out of 40)"))
  if marks < 16:
    print("fail")
    flag=1
    break
  # else:
    # print("all pass")
  i=i+1

if flag ==0:
  print("all pass")
```

```python
#method 3:: Python approach
i=0
while i<7:
  marks = int(input("Enter your marks (out of 40)"))
  if marks < 16:
    print("fail")
    break
  i=i+1
else: # while-else :: when no break is executed in while loop
  print("all pass")
```
