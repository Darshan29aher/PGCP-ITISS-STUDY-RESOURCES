```python
#WAP to find a number in given line
line1="this is my number 5678 or you can use 345"
regex="[0-9]+"
print(re.search(regex, line1))
print(re.findall(regex,line1))
```

```python
#WAP to find a mobile number in given line
line1="this is my number 1234565678 or you can use 345"
regex="[1-9][0-9]{9}"
print(re.search(regex, line1))
print(re.findall(regex,line1))
```

```python
#WA regex to match ip address from a line
regex_ip="[0-9]{1,3}.[0-9]{1,3}.[0-9]{1,3}.[0-9]{1,3}"
s1="10.10.10.10   192.168.7.12  1.1.1.1  255.255.255.255"
print(re.findall(regex_ip,s1))
```

# File handling

- open
open file in write , read or append mode

we can open file in binary write, read and append mode


- close


- read
read all contents of file as a single string

- write
write a given string into the file

## NOTE: for safe open and close we use with operator


```python
#create a temp.txt
with open("temp.txt","w") as f:
  print("file created!!")
#with by default closes the file safely, we don't need to do it
```

```python
#write a message "Hello" in file
with open("temp.txt","w") as f:
  f.write("Hello")
```

```python
#read contents of the file
with open("temp.txt","r") as f:
  contents = f.read()
print(contents)
```

```python
#read contents of the file
with open("/content/sample_log.txt","r") as f:
  contents = f.read()
for line in contents.split("\n"):
  print(line)
```
