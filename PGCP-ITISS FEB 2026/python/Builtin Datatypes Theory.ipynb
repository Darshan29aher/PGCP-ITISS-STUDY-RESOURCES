# What is immutable data type ?

**Immutable data type** : when we assign value then it **cannot be changed**.

**In case there is any modification** then new memory location is allocated.

**Example**: below diagram shows program temp1.py

**It is creating variable i with value 10.** **i is object reference on the stack. 10 is allocated memory on heap. When program assigns new value 20, then new memory location in heap is allocated.**

NOTE: Run code given below and you will see two different heap address (id) for the same variable, because we have changed its value in the program.

image.png

```python
# Program for Checking Immutable data type
# integer is immutable
i=10
print("i=",i,id(i),type(i)) # print id (means heap address of i) and data type of i

#now change value of i
i=20
print("i=",i,id(i),type(i)) # print id (means heap address) of i and data type of i
# Observe changes in the id (means heap address) of i
```

## python indexing --> +ve indexes and -ve indexes

- +ve index means 0,1,2...

- -ve indexes --> -1,-2, ...

-ve indexes start at the last element in the list / sequence , so last element is -1 index, second last -2, third last -3,

ex ::

l1 = [ 34 ,55 ,78 ,12,100,344,123 ]

negetive index means --> len(l1)-value

Ex : -5 -ve index --> len(l1)-5 --> 7 - 5 --> 2

So l1[-5] = l1[2] = 78

Always PVM converts negative indexes to positive indexes and then applies slicing





```python
l1=[10,20,30,40,50,60,70]
print(l1)
print(l1[-2]) #60 --> l1[7-2] --> l1[5] --> 60
print(l1[-5 : -2]) #start = -5 , end = -2, step = 1
# pvm converts -ve indexes to +ve --> start =7-5=2 end = 7-2=5 step=1
# pvm will see it as l1[2:5:1] -->
```

# Types of Copy in python

- shallow copy

it copies the reference from the list

here when we change immutable elements in copy then original remains as it is

here when we change mutable elements then they are changed in both original and copy

we use copy() function or slicing for creating shallow copy of list

- copy by assignment

here only reference of the list in stack is copied

so any changes in copy will affect the original

no extra memory on heap is allocated here

such copy is created when we use assignment operator

ex l1=[10,20,30]

l2 = l1 # --> copy by assignment


- deep copy ( using copy library)

here all mutable objects are allocated seperate memory on heap

so, **any changes** mutable / immutable to copy **will NOT affect the original**

it requires very high amount of memory on heap because all mutable objects are allocated new memory

```python
# shallow copy
l1 = [10,20,30,[400,500]]
# index 0 --> 10, inex 1 --> 20 index 2 --> 30, index 3--> [400,500]
l2 = l1.copy() # shallow copy
l1[0] = 99
# l2 is not affected
print(l1,l2)
l1[3][0]=999 #update 400
#l2 will be affected
print(l1,l2)
```

# **Lexicographical ordering**

- it is pythons way to compare two sequences

- it is applicable to comparison of lists / strings / tuples

- **Algorithm of lexicographical ordering**

1. take two strings (list / tuple)

2. compare first element of first string to first element of second string if same , repeat for the next element.

3. If all the elements of both strings are SAME  --> then both strings are SAME ...

ex

s1="ABC"

s2="ABC"

Both are same !!


4. if two elements are not same ... then string with higher value of element is **greater** and other is **smaller**

ex

s1="ABC"

s2="AD"

A and A are same , but B and D are not same. SO string with D is greater ...

s1 < s2

ex

s1="QWERTY"

s2="QWERTZ"

s1 < s2 --> Y is smaller than Z




5. if all elements are same **BUT** one string finishes , then string with more no of characters is **greater**

ex

s1 = "QAZ"

s2 = "QA"

"QA" are match  but s2 finishes ... so s1 is greater

s1 > s2



#### NOTE :: SAME algorithm works for list / string / tuple


# Set Data Type

Like sets in algebra

only unique / non repeatitive elements are allowed

set doesnt maintain order of insertion of elements

set is mutable , can be changed

in set we have add, deleted update insert of the elements

ONLY immutable elements (hashable) are allowed

So, int / float,complex,bool, string,tuple,frozenset are allowed

NO OTHER data type (list / set / dictionary) are allowed inside set

indexing and slicing are not allowed in set ( because it doesn't maintain insertion order!

ORDER of elements is maintained by hash value of an element

## How set finds duplicate elements ?

### Common simple questions with dictionary

```python
# take a sentence /paragraph
# find frequency of each word in the sentence/paragraph
para="cat sat on the mat mat is for the cat sat mat cat rat rat"
#seperate all words / tokens --> list of words
# WRONG APPROACH
print(set(para)) # --> gives uniques characters and not words
# WRONG APPROACH
print(list(para)) # --> gives ALL characters and not words
#RIGHT APPROACH
word_list = para.split(" ")
print(word_list)
#unique words --> set on word_list
unique_words = set(word_list)
print(unique_words)

#Solution 1 -- efficient
word_count = {}
for word in word_list:
  if word in word_count:
    word_count[word] += 1 #increase frequency of word
  else:
    word_count[word] = 1 # add new word with freq 1
print(word_count)

#solution2 --> not very efficient
word_count = {}
unique_words = set(word_list)
for word in unique_words:
  word_count[word] = word_list.count(word)

print(word_count)
```

```python
"""
20. Write a Python program to print all distinct values in a dictionary.
Sample Data : [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"},
{"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]

Expected Output : Unique Values: {'S005', 'S002', 'S007', 'S001', 'S009'}
"""
l_d = [{"V":"S001"}, {"V": "S002"}, {"VI": "S001"}, {"VI": "S005"},
{"VII":"S005"}, {"V":"S009"},{"VIII":"S007"}]
unique_vals=[]
for d1 in l_d:
  print(d1)
  if len(d1) == 1:
    val = list(d1.values())[0]
    print(val)
    if val not in unique_vals:
      unique_vals.append(val)
print(unique_vals)
```
